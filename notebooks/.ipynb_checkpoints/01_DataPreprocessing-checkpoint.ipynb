```python
# Rename useful columns with more brevity
df.rename(columns = {'round':'t_round','best_of':'m_best_of','match_num':'m_num','tourney_id':'t_id','tourney_name':'t_name','surface':'t_surf','draw_size':'t_draw_size','tourney_level':'t_lvl','tourney_date':'t_date','winner_id':'w_id','winner_seed':'w_sd','winner_entry':'w_ent','winner_name':'w_name','winner_ht':'w_ht','winner_hand':'w_hd','winner_ioc':'w_ioc','winner_age':'w_age', 'loser_id':'l_id','loser_seed':'l_sd','loser_entry':'l_ent','loser_name':'l_name','loser_hand':'l_hd','loser_ht':'l_ht','loser_ioc':'l_ioc','loser_age':'l_age','score':'m_score','minutes':'m_time(m)','winner_rank':'w_rank','winner_rank_points':'w_rank_pts','loser_rank':'l_rank','loser_rank_points':'l_rank_pts'}, inplace=True)
```

```python
df = df.drop(['w_df','w_1stIn','l_df','l_1stIn','w_sd','l_sd'],axis=1)
```

```python
# Remove Davis Cup and Grass Court tournaments, and matches resulting in Walkover. Also, restricting to only matches where at least 12 games were completed.
df = df[~df['t_name'].str.contains("Olympics")]
df = df[~df['t_name'].str.contains("Davis Cup")]
df = df[~df['t_surf'].str.contains("Grass")]
df = df[~df['m_score'].str.contains("W/O")]
df = df[(df['w_SvGms'] + df['l_SvGms'] >= 12)]
```

```python
df.info()
```

```python
df = df[["t_id", "t_surf", "tour_yr", "w_id", "w_name", "w_ht", "w_age", "l_id", "l_name", "l_ht", "l_age", "w_ace", "w_svpt", "w_1stWon", "w_2ndWon", "w_SvGms", "w_bpSaved", "w_bpFaced", "l_ace", "l_svpt", "l_1stWon", "l_2ndWon", "l_SvGms", "l_bpSaved", "l_bpFaced", "w_rank", "l_rank"]]
```

```python
df.info()
```

```python
# Ages reported with too much precision in the original file
df["w_age"] = df["w_age"].round(2)
df["l_age"] = df["l_age"].round(2)
```

```python
# Creation of target feature for each player in a given match: proportionalizing points played in the match appropriately between the two players.
df["w_pts_won%"] = round(((df["w_1stWon"] + df["w_2ndWon"] + (df["l_svpt"] - (df["l_1stWon"] + df["l_2ndWon"])))/ (df["w_svpt"] + df["l_svpt"]))*100, 2)

# Loser % pts won is simply 100 - w % pts won
df["l_pts_won%"] = 100 - df["w_pts_won%"]
```

```python
# Target broken down into serving and returning 

# Winner % Serve pts won
df["w_sv_pts_won%"] = round((df["w_1stWon"] + df["w_2ndWon"]) / df["w_svpt"]*100,2)

# Winner % Return pts won
df["w_ret_pts_won%"] =round(((df["l_svpt"] - (df["l_1stWon"] + df["l_2ndWon"]))/df["l_svpt"])*100,2)

# Loser % Serve pts won
df["l_sv_pts_won%"] = round((df["l_1stWon"] + df["l_2ndWon"]) / df["l_svpt"]*100,2)

# Loser % Return pts won
df["l_ret_pts_won%"] =round(((df["w_svpt"] - (df["w_1stWon"] + df["w_2ndWon"]))/df["w_svpt"])*100,2)
```

```python
# Not a target feature, but useful for generating predictive features: Total Points Played In Match
df["tot_pts"] = df["l_svpt"] + df["w_svpt"]
```

```python
df_winners = df.drop(["l_name", "l_rank", "l_ht", "l_age", "l_1stWon", "l_2ndWon", "l_SvGms", "l_pts_won%", "l_sv_pts_won%", "l_ret_pts_won%"], axis = 1)
df_winners["m_outcome"] = 1
df_losers =  df.drop(["w_name", "w_rank", "w_ht", "w_age", "w_1stWon", "w_2ndWon", "w_SvGms", "w_pts_won%", "w_sv_pts_won%", "w_ret_pts_won%"], axis = 1)
df_losers["m_outcome"] = 0
```

```python
# Split out winners and losers from by-match organization and concatenate into a per player organization
df_winners = df_winners.set_axis(["t_id", "t_surf", "tour_yr", "p_id", "p_name", "p_ht", "p_age", "opp_id", "p_ace", "p_svpt", "p_1stWon", "p_2ndWon", "p_SvGms", "p_bpSaved", "p_bpFaced", "opp_ace", "opp_svpt", "opp_bpSaved", "opp_bpFaced", "p_rank", "p_pts_won%", "p_sv_pts_won%", "p_ret_pts_won%", "tot_pts", "m_outcome"], axis=1)
df_losers = df_losers.set_axis(["t_id", "t_surf", "tour_yr", "opp_id", "p_id", "p_name", "p_ht", "p_age", "opp_ace", "opp_svpt", "opp_bpSaved", "opp_bpFaced", "p_ace", "p_svpt", "p_1stWon", "p_2ndWon", "p_SvGms", "p_bpSaved", "p_bpFaced", "p_rank", "p_pts_won%", "p_sv_pts_won%", "p_ret_pts_won%", "tot_pts", "m_outcome"], axis=1)
df_player1 = pd.concat([df_winners, df_losers], ignore_index=True)
df_player1 = df_player1.sort_values(by=['p_id','tour_yr'], ascending = False)
```

```python
# Filter down to only players in top 200 at any given time step (ie, match occurance)
df_player1 = df_player1.loc[(df_player1["p_rank"] <= 200)]
```

```python
# calculating player ace% per match 
df_player1["p_ace%"] = ((df_player1["p_ace"]/df_player1["p_svpt"])*100).round(2)
```

```python
# calculating player aced% per match 
df_player1["p_aced%"] = ((df_player1["opp_ace"]/df_player1["opp_svpt"])*100).round(2)
```

```python
# calculating player break point save% (as a server) per match
df_player1["p_bp_save%"] = ((df_player1["p_bpSaved"]/df_player1["p_bpFaced"])*100).round(2)
df_player1['p_bp_save%'] = df_player1['p_bp_save%'].fillna(100) #covers the cases where a player faced 0 break pts in the match
```

```python
# calculating player break point convert% (as a returner) per match
df_player1["p_bp_convert%"] = ((1 - (df_player1["opp_bpSaved"]/df_player1["opp_bpFaced"]))*100).round(2)
df_player1['p_bp_convert%'] = df_player1['p_bp_convert%'].fillna(0) #covers the cases where a player generated 0 break point opportunities in a match
```
